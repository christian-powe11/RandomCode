```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(0)

# Number of data points
n = 100

# Generate random data for the x and y coordinates
x = np.random.rand(n)  # Random values in a given shape [0, 1).
y = np.random.rand(n)  # Random values in a given shape [0, 1).

# Calculate the coefficients for the line of best fit
# Degree 1 means a linear function: y = ax + b
a, b = np.polyfit(x, y, 1)

# Generate y-values for the line of best fit
y_fit = a * x + b

# Create a scatter plot of the original data and a plot of the line of best fit
plt.scatter(x, y, label='Data Points')  # A scatter plot of y vs. x with varying marker size and/or color.
plt.plot(x, y_fit, color='red', label='Line of Best Fit')  # Line of best fit

# Adding necessary details to the plot
plt.title('Scatterplot with Line of Best Fit')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()  # Add a legend to describe elements of the plot

# Show the plot
plt.show()
```
