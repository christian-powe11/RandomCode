```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random data for linear regression
np.random.seed(42)  # for reproducibility
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Introducing an outlier and a high-leverage point
X = np.vstack([X, [[0.5], [2.5]]])  # x-values
y = np.vstack([y, [[12], [4]]])  # y-values. The first one is an outlier, the second is a high leverage point.

# Fitting a simple linear regression model
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X, y)
predictions = reg.predict(X)

# Plotting
plt.scatter(X, y, color="blue", s=50, label="Data points")
plt.plot(X, predictions, color="red", label="Regression line")
plt.scatter(0.5, 12, color="green", s=100, zorder=5, label="outlier")
plt.scatter(3, 8, color="red", s=100, zorder=5, label="leverage")
plt.title("Outliers and Leverage Points")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data: Age vs. Salary
ages = np.sort(np.random.randint(20, 60, 50))
salaries = 1000 + 50 * ages + np.random.normal(0, 300, len(ages))

# Gaussian kernel function
def gaussian_kernel(z):
    return (1/np.sqrt(2*np.pi)) * np.exp(-0.5 * z**2)

# Kernel regression
def kernel_regression(x, X, Y, bandwidth=5.0):
    weights = gaussian_kernel((x - X) / bandwidth)
    return np.sum(weights * Y) / np.sum(weights)

# Estimate salaries using kernel regression
predicted_salaries = [kernel_regression(age, ages, salaries) for age in ages]

# Plot
plt.scatter(ages, salaries, color='blue', label='Data points')
plt.plot(ages, predicted_salaries, color='red', label='Kernel regression')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.title('Kernel Regression: Age vs. Salary')
plt.show()
```
