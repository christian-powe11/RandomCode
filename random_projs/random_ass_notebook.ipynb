```python
import numpy as np

# Initialize lists to keep track of games and their lengths
games = []
lengths = []

# Run the simulation 99 times
for i in range(1, 100000):
    print(f"Running game {i}")
    
    # Initialize variables for a single game
    counter = 0
    game_in_progress = True
    games.append(i)
    while game_in_progress:
        # Roll three dice
        trial = np.random.randint(1, 7, 3)
        
        # Check if the sum is 18
        if sum(trial) == 18:
            game_in_progress = False
            lengths.append(counter)
        else:
            counter += 1
    

# Append the results for the last game

print("Games:", games)
print("Lengths:", lengths)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_games_vs_rounds(game_numbers, rounds_lasted):
    # Count the frequency of each number of rounds
    round_count = Counter(rounds_lasted)
    
    # Sort the rounds for better visualization
    sorted_rounds = sorted(round_count.keys())
    sorted_counts = [round_count[r] for r in sorted_rounds]
    
    # Plotting
    plt.bar(sorted_rounds, sorted_counts)
    
    # Adding labels and title
    plt.xlabel('Number of Rounds')
    plt.ylabel('Count of Games')
    plt.title('Count of Games vs Number of Rounds')
    
    # Show the plot
    plt.show()

# Sample data
game_numbers = games
rounds_lasted = lengths

# Call the function
plot_games_vs_rounds(game_numbers, rounds_lasted)
```
